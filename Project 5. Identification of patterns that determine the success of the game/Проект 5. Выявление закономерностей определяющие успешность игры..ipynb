# Предобработка данных:

**Приведение столбцов к нижнему регистру:**

```python
data.columns = data.columns.str.lower() # приведем все столбцы к нижнему регистру
data.columns # посмотрим на результат
```

**Работа со столбцом year_of_release:**

```python
data.dropna(subset = ['name', 'year_of_release', 'genre'], inplace = True) # удаляем пропущенные значения столбцов name, year_of_release, genre
data['year_of_release'] = data['year_of_release'].astype(int) # изменим тип данных столбца на int
```

**Работа со столбцом user_score:**

```python
data.loc[data['user_score'] == 'tbd', 'user_score'] = np.nan # заменеим пока значение "tbd" на nan
data['user_score'] = data['user_score'].astype('float') # изменим тип данных столбца на float
data['user_score'] = data['user_score'].fillna(-1) # заменим пропущенные значения в столбце 'user_score' значением -1
```

**Работа со столбцом critic_score:**

```python
data['critic_score'] = data['critic_score'].fillna(-1) # заменим пропущенные значения в столбце 'critic_score' значением -1
```

**Работа со столбцом rating:**

```python
data['rating'] = data['rating'].fillna('рейтинг неопределен') # заменим пропущенные значения в столбце 'rating' на 'рейтинг неопределен'
```

**Подсчет суммарных продаж во всех регионах и запись их в отдельный столбец:**

```python
data['sum_sales_games'] =  data[['na_sales', 'eu_sales', 'jp_sales', 'other_sales']].sum(axis = 1)
data.head() # выведем результат на экран
```

# Проверка гипотезы:

**Гиппотеза, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые:**

Имеем гипотезу, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые. Так как нам не важно в какую сторону проверяется значение рейтингов в большую или меньшую, то получается, что тест - двухсторонний. Применю специальный метод для проверки гипотезы: который говорит о равенстве среднего двух генеральных совокупностей по взятым из них выборкам: scipy.stats.ttest_ind(array1, array2, equal_var). equal_var возьмем равный False, а параметр alpha (критический уровень статистической значимости) = 0.05. Важно учесть, что нулевая гипотеза не может предполагать неравенство. Условие ее применения - это отсутствие различий, отсутствие влияния какого-либо фактора, но при этом она допускает наличие незначительных, очень малых случайных значений. Исходя из этого гипотезы зададим следующим образом:

Для H0 (нулевая гипотеза) возьмем, что средние пользовательские рейтинги платформ Xbox One и PC равные;

Для H1 (альтернативная гипотеза) возьмем, что средние пользовательские рейтинги платформ Xbox One и PC разные (не равны).

```python
from scipy import stats as st
```

```python
results = st.ttest_ind(
    actual_data[actual_data['platform'] == 'XOne'] ['user_score'],
    actual_data[actual_data['platform'] == 'PC'] ['user_score'],
    equal_var = False)

alpha = 0.05 # значение уровня значимости

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print(results.pvalue, "Отвергаем нулевую гипотезу")
else:
    print(results.pvalue, "Не получилось отвергнуть нулевую гипотезу") 
```

p-value показывает, что вероятность случайно получить такое или большее различие равно почти 12 процентов, это очевидно, довольно большая вероятность, чтобы сделать вывод о значимом различии между средними пользовательскими рейтингами платформ XOne и PC. Соответственно не получилось отвергнуть нулевую гипотезу.
